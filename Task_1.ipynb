```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = '/content/SPI_index_labelled.csv'

# Read the csv file into a DataFrame
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
# Filter out NaN values if needed
data = df['region'].dropna()
```

###Create the histogram of region

```python
fig, ax = plt.subplots(figsize=(15, 5))

ax.hist(data, bins=10, color='skyblue', edgecolor='black')

ax.set_xlabel('region')
ax.set_ylabel('Frequency')
ax.set_title('Histogram of region')

plt.tight_layout()
plt.show()
```
